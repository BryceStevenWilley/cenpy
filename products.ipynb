```python
from cenpy import products
import pandas
import numpy
import matplotlib.pyplot as plt
import cenpy
import requests
%matplotlib inline
```

```python
dectest = products.Decennial2010()
```

```python
la = dectest.from_county('Los Angeles, CA', 
                          level='tract', 
                          variables=['^P004'])
```

```python
f,ax = plt.subplots(1,1,figsize=(20,20))
la.plot('P004003', edgecolor='white', cmap='plasma', ax=ax, linewidth=.2)
ax.set_facecolor('k')
```

```python
aus, bounds_aus = dectest.from_msa('Austin, TX', 
                                   level='block', 
                                   variables=['P003003'], 
                                   return_bounds=True)
```

```python
f,ax = plt.subplots(1,1,figsize=(20,20))
aus.plot('P003003',cmap='plasma', linewidth=0, ax=ax)
bounds_aus.plot(ax=ax, edgecolor='slategray', facecolor='none', linewidth=2)
ax.set_facecolor('k')
```

```python
ks, ks_bounds = dectest.from_csa('Kansas City', 
                                 level='tract', 
                                 variables=['P001001'],
                                 return_bounds=True)
```

```python
f,ax = plt.subplots(1,1,figsize=(20,20))
ks.plot('P001001', ax=ax, cmap='plasma', linewidth=0)
ks_bounds.plot(ax=ax, edgecolor='w', linewidth=2, color='none')
ax.set_facecolor('k')
```

```python
ma, ma_bounds = dectest.from_state('Massachusetts', 
                                   strict_within=False, 
                                   return_bounds=True)
```

```python
f, ax = plt.subplots(1,1,figsize=(20,20))
ma.plot(edgecolor='slategrey', color='none', ax=ax)
ma_bounds.plot(edgecolor='cornsilk', color='none', ax=ax, linewidth=2)
ax.set_facecolor('k')
```

```python
tuc, tuc_bounds = dectest.from_place('Tucson, AZ', 
                                     level='tract',
                                     return_bounds=True, 
                                     strict_within=False)
```

```python
f, ax = plt.subplots(1,1,figsize=(20,20))
tuc.plot(edgecolor='slategrey', color='none', ax=ax)
tuc_bounds.plot(edgecolor='cornsilk', color='none', ax=ax, linewidth=2)
ax.set_facecolor('k')
```
